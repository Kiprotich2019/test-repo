## Summary statistics

```python
sales=pd.read_csv("sales_subset.csv", index_col="Unnamed: 0")
sales
sales.head()
#sales.info()
sales["weekly_sales"].mean()
sales["weekly_sales"].median()
sales["date"].max()
sales["date"].min()
```

### Efficient Summaries

```python
def iqr(column):
    return column.quantile(0.75)-column.quantile(0.25)
sales[["temperature_c","fuel_price_usd_per_l","unemployment"]].agg(iqr)
sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg([iqr,np.median])
```

## Cumulative Statistics

```python
sales
sales_1_1=sales[(sales["store"]==1) & (sales["department"]==1)]
sales_1_1.sort_values("date")
sales_1_1["cum_weekly_sales"]= sales_1_1["weekly_sales"].cumsum()
```

```python
sales_1_1["cum_max_sales"]=sales_1_1["weekly_sales"].cummax()
sales_1_1
```

## Counting

### Dropping duplicate names

```python
sales
store_types=sales.drop_duplicates(subset=["type", "store"])
store_types.head()
store_depts=sales.drop_duplicates(subset=["store", "department"])
store_depts.head()
holiday=sales[sales["is_holiday"]== True]
holiday_dates=holiday.drop_duplicates(subset="date")
holiday_dates["date"]
```

### Counting values

```python
store_types["type"].value_counts()
store_types["type"].value_counts(normalize= True)
store_depts["department"].value_counts(sort= True)
store_depts["department"].value_counts(sort=True, normalize=True)
```

## Grouped Summary Statistics

```python
sales_all=sales['weekly_sales'].sum()
sales_A=sales[sales["type"]== "A"]["weekly_sales"].sum()
sales_B = sales[sales["type"] == "B"]["weekly_sales"].sum()
sales_C = sales[sales["type"] == "C"]["weekly_sales"].sum()
sales_propn_by_type = [sales_A, sales_B, sales_C] / sales_all
sales_propn_by_type
sales_by_type=sales.groupby("type")["weekly_sales"].sum()
sales_propn_by_type = sales_by_type / sum(sales_by_type)
sales_propn_by_type
```

## Multiple Group Summaries

```python
sales_stats=sales.groupby("type")["weekly_sales"].agg([max,min,np.mean,np.median])
sales_stats
unemp_fuel_stats = sales.groupby('type')[['unemployment',"fuel_price_usd_per_l"]].agg([min, max, np.mean, np.median])
unemp_fuel_stats
```

## Pivot tables

## Simple regression

```python
data=pd.read_csv("8.3 1.01. Simple linear regression.csv")
data.describe()
```

## define dependent and independent variables

```python
y=data["GPA"]
x1=data["SAT"]
#data.plot(x='SAT',y='GPA',kind="scatter")
plt.show()
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat=0.0017*x1+0.2750
fig=plt.plot(x1,yhat,lw=4,c='black',label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```
